```python
df
```

```python
for row in df.iterrows():
  print(row)
```

```python
marathi_tweets=[]
hindi_tweets=[]
other_tweets=[]
for col in df.columns:
   #print(col)
   sen = df[col]
   print(type(sen))
   for each in sen:
     print(each)
     lang = cld3.get_language(str(each))
     #print(lang.language)
     #print(type(lang))
     if lang.language =='mr' and lang.probability>0.8:
       marathi_tweets.append(each)
     elif lang.language == 'hi'and lang.probability>0.8:
       hindi_tweets.append(each)
     else:
       other_tweets.append(each)
      
```

```python
max_length = max(max(len(marathi_tweets), len(hindi_tweets)), len(other_tweets))
marathi_tweets += ['Nan'] * (max_length - len(marathi_tweets))
hindi_tweets += ['Nan'] * (max_length - len(hindi_tweets))
other_tweets += ['Nan'] * (max_length - len(other_tweets))
```

```python
print("Marathi tweets number:",len(marathi_tweets))
print("Hindi tweets number:",len(hindi_tweets))
print("Other tweets number:",len(other_tweets))
```

```python
dict = {"marathi tweets": marathi_tweets, "hindi tweets": hindi_tweets,"other tweets": other_tweets}
df = pd.DataFrame(dict)

df.to_csv('shows.csv')
```
